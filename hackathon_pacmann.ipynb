```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("/content/tourism_rating.csv")
df2 = pd.read_csv("/content/tourism_with_id.csv")
df3 = pd.read_csv("/content/user.csv")
```

```python
df1.sample(5)
```

```python
df2.sample(5)
```

```python
df3.sample(5)
```

```python
raw = pd.merge(df1, df3, on='User_Id')
raw.sample(2)
```

```python
df = pd.merge(df2, raw, on='Place_Id')
```

```python
df.sample(5)
```

```python
df = df.drop(columns=['Unnamed: 11', 'Unnamed: 12'])
```

```python
df.head()
```

```python
df = df.rename(columns={'Location':'From'})
```

```python
df
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df["Age"], bins=20, edgecolor="black")
plt.title("User Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
avg_rating_by_category = df.groupby("Category")["Rating"].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
avg_rating_by_category.plot(kind="bar")
plt.title("Average Rating by Category")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df["Rating"], bins=10, edgecolor="black")
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["Time_Minutes"], df["Rating"])
plt.title("Rating vs. Time Spent")
plt.xlabel("Time (Minutes)")
plt.ylabel("Rating")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["Age"], df["Price"])
plt.title("User Age vs. Price")
plt.xlabel("User Age")
plt.ylabel("Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
categories = df["Category"].unique()
num_categories = len(categories)

fig, axes = plt.subplots(nrows=num_categories, ncols=1, figsize=(8, 4 * num_categories))

for idx, category in enumerate(categories):
    sns.histplot(df[df["Category"] == category]["Age"], bins=20, ax=axes[idx], kde=True)
    axes[idx].set_title(f"Age Distribution for Category: {category}")
    axes[idx].set_xlabel("Age")
    axes[idx].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Price", y="Rating")
plt.title("Price vs rating by Category")
plt.xlabel("Price")
plt.ylabel("Rating")
plt.legend(title="Category")
plt.grid(True)
plt.tight_layout()
plt.show()
```
